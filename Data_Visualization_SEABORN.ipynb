# LINEPLOT

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",style="sex")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm" , y="bill_depth_mm" , data=df , hue="sex" , palette="Accent")
plt.show()
```

```python
sns.lineplot(x="bill_length_mm",y="bill_depth_mm",data=df,hue="sex",palette="Accent",markers=["o",">"])
plt.show()
```

# FACET-GRID:

```python
df1 = sns.load_dataset("tips")
df1.head()
```

```python
fg = sns.FacetGrid(df1,col="sex",hue="day")
fg.map(plt.scatter , "total_bill" ,"tip").add_legend()
plt.show()
 
```

```python
fg = sns.FacetGrid(df1,col="smoker",hue="sex")
fg.map(plt.bar , "total_bill" ,"tip").add_legend()
plt.show()
 
```

```python
fg = sns.FacetGrid(df1,col="day",hue="sex")
fg.map(plt.scatter , "total_bill" ,"size").add_legend()
plt.show()
 
```

```python
fg = sns.FacetGrid(df1,col="smoker",hue="sex",palette="summer")
fg.map(plt.bar , "total_bill" ,"tip").add_legend()
plt.show()
 
```

```python
var = sns.load_dataset("penguins")
```

```python
var.head()
```

```python
sns.barplot(x=var.island,y=var.bill_length_mm)
plt.show()
```

```python
sns.barplot(data=var , x="island",y="bill_length_mm" , hue="sex")
plt.show()
```

```python
## WE CAN MANIPULATE THE ORDER OF EACH GRAPH:
order_1 = ["Biscoe",'Dream','Torgersen']
sns.barplot(data=var , x="island",y="bill_length_mm" , hue="sex",order=order_1,palette="summer")
plt.show()
```

```python
order_1 = ["Biscoe",'Dream','Torgersen']
sns.barplot(data=var , x="island",y="bill_length_mm" , hue="sex",order=order_1,palette="spring",errcolor='black',capsize=0.3)
plt.grid()
plt.show()
```
